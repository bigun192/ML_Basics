```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file_path = "C:\Python\python-data-analysis-master\data\market-price.csv"
bitcoin_df = pd.read_csv(file_path, names = ['day','price'])


bitcoin_df.head()
# 시계열 비트코인 정보를 데이터 프레임의 인덱스로 설정하여 가격의 추이를 시각화 

#to_datetime 으로 day 피처를 시계열 피처로 변환합니다.
#bitcoin_df.head()
bitcoin_df['day'] = pd.to_datetime(bitcoin_df['day'])
bitcoin_df['day'].dt.month

#day 데이터 프레임의 index로 설정합니다
bitcoin_df.index = bitcoin_df['day']
bitcoin_df.set_index('day', inplace = True)
bitcoin_df.index
# 일자별 비트코인 시세를 시각화
#bitcoin_df.plot()
#plt.show()

#ARIMA 모델 활용하기 : 모델학습
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm

#(AR = 2, 차분 = 1, MA = 2) 파라미터로 ARIMA 모델을 학습

model = ARIMA(bitcoin_df.price.values, order = (2,1,2))
model_fit = model.fit(trend = 'c', full_output= True, disp = True)
#print(model_fit.summary())

#ARIMA 모델 활용하기: 모델의 성능 & 예측결과 시각화
#fig = model_fit.plot_predict() # 학습 데이터에 대한 예측 결과  
                            #학습한 모델에 학습 데이터 셋을 넣었을 때의 시계열 예측 결과. plot_predict()함수로 시각화를 자동으로 수행
#residuals = pd.DataFrame(model_fit.resid) # 잔차의 변동을 시각화합니다(두번째 그래프)
#residuals.plot()

#ARIMA 모델 평가

'''
모델평가 과정
1. model_fit.forecast(steps =5) 로 향후 5일간의 가격을 예측
2. "C:\Python\python-data-analysis-master\data\market-price.csv"에서 실제 향후 5일의 가격을 test_y로 정의
3. 모델이 예측한 상한값, 하한값을 pred_y_upper, pred_y_lower로 정의
4. 정의한 모든 값을 비교하여 5일 동안의 상승 경향 예측이 얼마나 맞았는지를 평가한다

'''

forecast_Data = model_fit.forecast(steps = 5) # 학습 데이터 셋으로부터 5일뒤를 예측합니다

#테스트 데이터 셋을 불러옵니다
test_df = pd.read_csv("C:\Python\python-data-analysis-master\data\market-price-test.csv", names= ['ds', 'y'])

pred_y = forecast_Data[0].tolist() # 마지막 5일의 예측 데이터입니다.
test_y = test_df.y.values # 실제 5일 가격 데이터입니다
#print(test_y)
pred_y_lower = [] #마지막 5일의 예측 데이터의 최소값입니다.
pred_y_upper = [] # 마지막 5일의 예측 데이터의 최대값입니다.

for lower_upper in forecast_Data[2]:
    lower = lower_upper[0]
    upper = lower_upper[1]
    pred_y_lower.append(lower)
    pred_y_upper.append(upper)
    
# 위의 코드를 시각화

plt.plot(pred_y , color = "gold")
plt.plot(pred_y_lower,color = "red")
plt.plot(pred_y_upper , color = "blue")
plt.plot(test_y,color = "green")


```
